 # regression model

```python
df.head()
```

```python
df.corr()
```

```python
# splitting the dataframe into two dfs

x = df.drop('price_new',axis=1)
x1=pd.DataFrame(x)
print(x1)
y = df['price_new']
y1 = pd.DataFrame(y)
print(y1)
```

```python
# now we got x1 and y1
```

```python
#we will split both variables into the test set and training set.
#We are splitting our dataset so that we can train our model using a training dataset and then test the model using a test dataset.

# Splitting the dataset into training and test set.

from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test = train_test_split(x1,y1,test_size=0.25,random_state=42)
```

```python
# now we do the train and test on the datasets

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
# we called the library and created a instance
```

```python
# creatin a model on the training set
model=lr.fit(x1_train,y1_train)
```

```python
# now we use this model onto the test
```

```python
y1_pred = model.predict(x1_test)

print(y1_pred)
```

```python
print(y1_test)
```

```python
# now we can check the accuracy by two ways
```

```python
# mean_squared_error

from sklearn.metrics import mean_squared_error
print('mse is:',mean_squared_error(y1_test,y1_pred))
```

```python
# r2_score

from sklearn.metrics import r2_score
print("r2 score:",r2_score(y1_test,y1_pred))
```
