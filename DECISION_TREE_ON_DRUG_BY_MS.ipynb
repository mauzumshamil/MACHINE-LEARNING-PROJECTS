```python
x = df.drop('Drug',axis=1)
x1 = pd.DataFrame(x)
y = df['Drug']
y1 = pd.DataFrame(y)
```

```python
model = DTC.fit(x1_train,y1_train)
```

```python
y1_pred = model.predict(x1_test)
```
