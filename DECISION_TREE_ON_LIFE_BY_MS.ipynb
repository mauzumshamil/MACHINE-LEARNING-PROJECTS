```python
columns = df[['Life expectancy ', 'Adult Mortality', 'Alcohol',
           'Hepatitis B', ' BMI ', 'Polio', 'Total expenditure',
           'Diphtheria ', 'GDP', 'Population', ' thinness  1-19 years', ' thinness 5-9 years',
           'Income composition of resources', 'Schooling']]

# Create subplots
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(20, 20))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Iterate through columns and plot histograms
for i, column in enumerate(columns):
    ax = axes[i]
    ax.hist(df[column].dropna(), bins=20, color='skyblue', edgecolor='black')  # Adjust the number of bins as needed
    ax.set_title(column)
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# DECISION TREE WITH REGRESSOR
```

```python
 x = df.drop('Life expectancy ',axis=1)
 x1 = pd.DataFrame(x)
 y = df['Life expectancy ']
 y1 = pd.DataFrame(y)
```

```python
from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test= train_test_split(x1,y1,test_size=0.25,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
DTR = DecisionTreeRegressor(random_state=42)

model = DTR.fit(x1_train,y1_train)
y1_pred= model.predict(x1_test)
```

```python
print(y1_test)
print(y1_pred)
```

```python
# validation with mse and r2 score
```

```python
from sklearn.metrics import mean_squared_error
print("the mse is:", mean_squared_error(y1_test,y1_pred))
```

```python
from sklearn.metrics import r2_score
print("the r2 score is:", r2_score(y1_test,y1_pred))
```
